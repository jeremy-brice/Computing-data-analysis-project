# $\textcolor{blue}{Method}$

### 1D wave equations
These coupled equations are used to model the behavior of waves in a shallow fluid, such as water in a canal or a shallow ocean: 

$$
\frac{\partial \eta}{\partial t} + H \frac{\partial u}{\partial x} = 0 \tag{1}
$$

$$
\frac{\partial u}{\partial t} = -g \frac{\partial \eta}{\partial x} \tag{2}
$$

where **η(x, t)** represents sea surface elevation relative to a fixed depth **H = 100m**, **u(x, t)** represents the velocity, and **$g = 9.81 m.s^{-2}$** is the gravitational constant. 

These equations can be modified to a single equation for η, it's the combined equation:  $\frac{\partial^2 \eta}{\partial t^2} = c^2 \frac{\partial^2 \eta}{\partial x^2}$ with $c = \sqrt{gH}$  
We will see through this assignment the 1D wave equation with different boundary conditions.

### 2D wave equations
$$
\frac{\partial \eta}{\partial t} + H \left( \frac{\partial u}{\partial x} + \frac{\partial v}{\partial y} \right) = 0 \tag{3}
$$
$$
\frac{\partial u}{\partial t} = -g \frac{\partial \eta}{\partial x} \tag{4}
$$
$$
\frac{\partial v}{\partial t} = -g \frac{\partial \eta}{\partial y} \tag{5}
$$ 


These equations can be modified to a single equation for η:  $\frac{\partial^2 \eta}{\partial t^2} = c^2 \left( \frac{\partial^2 \eta}{\partial x^2} + \frac{\partial^2 \eta}{\partial y^2} \right)
$  
We will see through this assignment the 2D wave equation with different boundary conditions.

### FTCS 
The FTCS is first order in time but second ordrer in space, so it's spatially more stable, it acts symmetrically around a point, so it's fit with diffusion that is isotrope. It's an easy scheme to implement and more precise.
$$
\frac{\eta_{i}^{n+1} - \eta_{i}^{n}}{\Delta t} + H \frac{u_{i+1}^{n} - u_{i-1}^{n}}{2\Delta x} = 0
$$

### Grid stateggering
We use the grid staggering to improve the precision and stability. The point is to shift the variable positions instead of having them all at the same point. That permits to better understand the gradiant, In time, we can offset the grid points of one variable in time to improve the performance of our numerical solutions.  (see **Sckecth 1** below)


### CFL
The FTCS scheme is an explicit one, which means that stability depends directly on the number of CFLs. We need CFL ≤ 1 with :  
$CFL = c \cdot \frac{\Delta x}{\Delta t} \leq 1$  
The propagation speed c can be estimated as $c = \sqrt{gH}$
In our general setup we have $ g=9.81 m.s ^{-1}$ the constant gravity, and we have H=100m. I will not change those two constants in this assignment.

# $\textcolor{blue}{General\ Set up}$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import imageio.v2 as imageio
from IPython.display import Image, display
from scipy.interpolate import interp1d
```

```python
#Time
Tsteps = 600
dt = 0.1 # timestep [seconds]

#Space
dx = 10 # space step [meters]
dy = 10 
Xpoints = 100
Ypoints = 100
xrange= np.linspace(0,Xpoints*dx, Xpoints)
yrange= np.linspace(0,Ypoints*dx, Ypoints)

# constants
H = 100   
g = 9.81  
```

```python
# initial condition ranges - where the gaussian bump is defined
start_x = 35
end_x = 65
start_y = 35
end_y = 65


def gaussian_1d(X, start_x, end_x):
    x = np.arange(X)
    mean = (start_x + end_x) / 2
    std_dev = (end_x - start_x) / 6
    return np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

def gaussian_2d (X, Y, start_x, end_x, start_y, end_y):
    x = np.arange(X)
    y = np.arange(Y)
    mean_x = (start_x + end_x) / 2
    std_dev_x = (end_x - start_x) / 6
    mean_y = (start_y + end_y) / 2
    std_dev_y = (end_y - start_y) / 6
    X, Y = np.meshgrid(x, y, indexing='ij')
    return np.exp(-0.5 * (((X - mean_x) / std_dev_x) ** 2 +
                           ((Y - mean_y) / std_dev_y) ** 2))
```

```python
eta_0_1d = gaussian_1d(Xpoints, start_x, end_x)
eta_0_2d = gaussian_2d(Xpoints, Ypoints, start_x, end_x, start_y, end_y)
u_0 = np.zeros(Xpoints)

plt.figure(figsize=(10, 4))
plt.plot(xrange, eta_0_1d, label="η t=0s")
plt.plot(xrange, u_0, label="u t=0s")
plt.xlabel("Position $x$ (m)")
plt.ylabel("Amplitude $η(x, 0)$" (m))
plt.title("Fig1 : Initial condition 1D")
plt.legend()
plt.xlim(0, 1000) 
plt.show()

plt.figure(figsize=(8, 6))
plt.imshow(eta_0_2d, extent=[0, Xpoints*dx, 0, Ypoints*dx], origin='lower', cmap='viridis')
plt.colorbar(label='Amplitude η')
plt.xlabel("Position x (m)")
plt.ylabel("Position y")
plt.title("Fig2 : Initial condition 2D")
plt.show()
```

**Figure 1** shows the initial condition in one dimension for eta and u. The sea surface elevation, eta, is a smooth bell-shaped curve. The speed, u, is zero. This setup represents a perturbation in the form of a wave. Since the speed is zero, the water is initially "calm," allowing us to focus on the speed generated by the perturbation easily.

**Figure 2** shows the amplitude eta at t=0, it is the initial condition. It is represented as a bell-shaped curve in two dimensions, viewed from above. The speeds u and v are also equal to zero for the same reasons as in the one-dimensional case.

# $\textcolor{blue}{Part1:\ Staggered\  FTCS\ scheme\ coupeld\ equations\ 1D}$
## Boundaries condition : wall for u and one step before for eta 

```python
# solution for time & space staggered FTCS

# derived constants
cg = g * dt / dx
cH = H * dt / dx

# initializing 
eta_FT_1d = np.zeros([Tsteps , Xpoints])
u_FT_1d  = np.zeros([Tsteps , Xpoints+1]) #one longer because of staggering

# apply initial condition
eta_FT_1d[0,:] = gaussian_1d(Xpoints, start_x, end_x)
```

```python
########### 1D ########################################################
for n in range(Tsteps-1):
    for i in range(1,Xpoints-1):     

        eta_FT_1d[n + 1, i] = eta_FT_1d[n, i] - cH * (u_FT_1d[n, i + 1] - u_FT_1d[n, i])
    
    # wall for U and one step before for eta 1D
    eta_FT_1d[n + 1, 0] = eta_FT_1d[n+1, 1]  
    eta_FT_1d[n + 1, -1] = eta_FT_1d[n+1, -2] 
    
    for i in range(1,Xpoints):  
        u_FT_1d[n + 1, i] = u_FT_1d[n, i] - cg * (eta_FT_1d[n + 1, i] - eta_FT_1d[n + 1, i - 1])

    #wall for U and one step before for eta 1D
    u_FT_1d[n + 1, 0] = u_FT_1d[n+1, 1]
    u_FT_1d[n + 1, -1] = u_FT_1d[n+1, 0]
```

eta_FT_1d[n + 1, 0] = eta_FT_1d[n+1, 1]  the value of eta at the left edge (i = 0) is equal to the value of eta at the adjacent point (i = 1)   
u_FT_1d[n + 1, -1] = u_FT_1d[n+1, 0] means that the value of u at the right edge (i = -1) is equal to the value of u at the left edge (i = 0).  
When eta have reflective boundary that mean there is no flux accross the boundary. 

```python
def interpolate(e,u):    #code given by teacher
    u_e= interp1d(np.linspace(0,1,u.shape[1]), u, axis=1)(np.linspace(0,1,e.shape[1]))
    #v_e= interp1d(np.linspace(0,1,v.shape[1]), v, axis=1)(np.linspace(0,1,e.shape[1]))

    return(u_e)
```

```python
#1D
# Time steps to plot
time_steps_to_plot = [0, 10, 490, 500]

#eta(x, t)
plt.figure(figsize=(10, 6))
for t in time_steps_to_plot:
    plt.plot(xrange, eta_FT_1d[t, :], label=f"η(x, {t * dt}s)")

plt.xlabel("Position $x$ (m)")
plt.ylabel("Amplitude η (m)")
plt.title("Fig3a : FTCS waves 1D - η(x, t) - BC = step before")
plt.legend()
plt.xlim(0, Xpoints * dx)
plt.grid(True)
plt.show()

# u(x, t)
u_e_1D = interpolate(eta_FT_1d, u_FT_1d)  #calling the interpolate u

plt.figure(figsize=(10, 6))
for t in time_steps_to_plot:
    plt.plot(xrange, u_e_1D[t, :], label=f"u(x, {t * dt}s)", linestyle='--')

plt.xlabel("Position $x$ (m)")
plt.ylabel("u")
plt.title("Fig3b : FTCS waves 1D - u(x, t) - BC = wall")
plt.legend()
plt.xlim(0, Xpoints * dx)
plt.grid(True)
plt.show()
```

## Von neumann wall boundaries with zero velocity gradient across the boundaries  
Boundary conditions of Von Neumann are given by: $\frac{\partial u(0, t)}{\partial x} = 0 \quad \text{et} \quad \frac{\partial u(L, t)}{\partial x} = 0$.  
The value of the derivative of the speed at the domain boundary is defined as 0, the gradient is 0. This means that the velocity at the boundary is equal to the velocity just inside the domain. In terms of discretization, this can be interpreted as: **u(0,t) = u(1,t) and u(L,t) = u(L-1,t)**. It represents something slippery outside of the domain. 

For eta, we can apply the same process, the boundary values will also be the same as the interior. we will have **η(0,t)=η(1,t) et η(L,t)=η(L−1,t)**. 

```python
# initializing 
eta_FT_1d_VN = np.zeros([Tsteps , Xpoints])
u_FT_1d_VN  = np.zeros([Tsteps , Xpoints+1]) #one longer because of staggering

# apply initial condition
eta_FT_1d_VN[0,:] = gaussian_1d(Xpoints, start_x, end_x)

############ Von neumann ########################################################
for n in range(Tsteps - 1):
    for i in range(1, Xpoints - 1):
        eta_FT_1d_VN[n + 1, i] = eta_FT_1d_VN[n, i] - cH * (u_FT_1d_VN[n, i + 1] - u_FT_1d_VN[n, i])

    # boundaries condition von Neumann  eta
    eta_FT_1d_VN[n + 1, 0] = eta_FT_1d_VN[n + 1, 1]
    eta_FT_1d_VN[n + 1, -1] = eta_FT_1d_VN[n + 1, -2]

    for i in range(1, Xpoints):
        u_FT_1d_VN[n + 1, i] = u_FT_1d_VN[n, i] - cg * (eta_FT_1d_VN[n + 1, i] - eta_FT_1d_VN[n + 1, i - 1])

    # boundaries condition von Neumann pour u
    u_FT_1d_VN[n + 1, 0] = u_FT_1d_VN[n + 1, 1]
    u_FT_1d_VN[n + 1, -1] = u_FT_1d_VN[n + 1, -2]
```

```python
# Time steps to plot
time_steps_to_plot = [0, 10, 490, 500]

#eta(x, t)
plt.figure(figsize=(10, 6))
for t in time_steps_to_plot:
    plt.plot(xrange, eta_FT_1d_VN[t, :], label=f"η(x, {t * dt}s)")

plt.xlabel("Position $x$")
plt.ylabel("Amplitude η")
plt.title("Fig4a : FTCS waves 1D - η(x, t) -with von Neumann boundary conditions")
plt.legend()
plt.xlim(0, Xpoints * dx)
plt.grid(True)
plt.show()

# u(x, t)
u_e_1D_VN = interpolate(eta_FT_1d_VN, u_FT_1d_VN)

plt.figure(figsize=(10, 6))
for t in time_steps_to_plot:
    plt.plot(xrange, u_e_1D_VN[t, :], label=f"u(x, {t * dt}s)", linestyle='--')

plt.xlabel("Position $x$")
plt.ylabel(" u")
plt.title("Fig4b : FTCS waves 1D - u(x, t) von Neumann boundary conditions")
plt.legend()
plt.xlim(0, Xpoints * dx)
plt.grid(True)
plt.show()
```

# $\textcolor{blue}{Numerical\ FTCS\ scheme\ coupeld\ equations\ 2D}$
## Periodic boundaries condition

```python
def interpolatev(e,v):    #code given
    #u_e= interp1d(np.linspace(0,1,u.shape[1]), u, axis=1)(np.linspace(0,1,e.shape[1]))
    v_e= interp1d(np.linspace(0,1,v.shape[1]), v, axis=1)(np.linspace(0,1,e.shape[1]))

    return(v_e)
```

```python
# solution for time & space staggered FTCS

# derived constants
cg = g * dt / dx
cH = H * dt / dx

# initializing 
eta_FT_2d = np.zeros((Tsteps, Xpoints, Ypoints))
u_FT_2d = np.zeros((Tsteps, Xpoints + 1, Ypoints))  # one longer because of staggering
v_FT_2d = np.zeros((Tsteps, Xpoints, Ypoints + 1))  # one longer because of staggering

# apply initial condition
eta_FT_2d[0, :, :] = gaussian_2d(Xpoints, Ypoints, start_x, end_x, start_y, end_y)

#interpolate
u_e_FT_2d = interpolate(eta_FT_2d, u_FT_2d)
v_e_FT_2d = interpolatev(eta_FT_2d, v_FT_2d)

```

```python
########## time-stepping loop 2D ###########################################
for n in range(Tsteps - 1):
    for i in range(1, Xpoints - 1):
        for j in range(1, Ypoints - 1):
            eta_FT_2d[n + 1, i, j] = eta_FT_2d[n, i, j] - cH * (u_e_FT_2d[n, i + 1, j] - u_e_FT_2d[n, i, j]) - cH * (v_e_FT_2d[n, i, j + 1] - v_e_FT_2d[n, i, j])
            u_e_FT_2d[n + 1, i, j] = u_e_FT_2d[n, i, j] - cg * (eta_FT_2d[n + 1, i, j] - eta_FT_2d[n + 1, i - 1, j])
            v_e_FT_2d[n + 1, i, j] = v_e_FT_2d[n, i, j] - cg * (eta_FT_2d[n + 1, i, j] - eta_FT_2d[n + 1, i, j - 1])

    # periodics boundary conditions 2D
    eta_FT_2d[n + 1, 0, :] = eta_FT_2d[n, 0, :] - cH * (u_e_FT_2d[n, 1, :] - u_e_FT_2d[n, 0, :])
    eta_FT_2d[n + 1, -1, :] = eta_FT_2d[n, -1, :] - cH * (u_e_FT_2d[n, 0, :] - u_e_FT_2d[n, -1, :])
    eta_FT_2d[n + 1, :, 0] = eta_FT_2d[n, :, 0] - cH * (v_e_FT_2d[n, :, 1] - v_e_FT_2d[n, :, 0])
    eta_FT_2d[n + 1, :, -1] = eta_FT_2d[n, :, -1] - cH * (v_e_FT_2d[n, :, 0] - v_e_FT_2d[n, :, -1])

    u_e_FT_2d[n + 1, 0, :] = u_e_FT_2d[n, 0, :] - cg * (eta_FT_2d[n + 1, 0, :] - eta_FT_2d[n + 1, -1, :])
    u_e_FT_2d[n + 1, -1, :] = u_e_FT_2d[n, -1, :] - cg * (eta_FT_2d[n + 1, -1, :] - eta_FT_2d[n + 1, -2, :])
    v_e_FT_2d[n + 1, :, 0] = v_e_FT_2d[n, :, 0] - cg * (eta_FT_2d[n + 1, :, 0] - eta_FT_2d[n + 1, :, -1])
    v_e_FT_2d[n + 1, :, -1] = v_e_FT_2d[n, :, -1] - cg * (eta_FT_2d[n + 1, :, -1] - eta_FT_2d[n + 1, :, -2])
```

```python
#2D
# Time steps to plot
time_steps_to_plot = [0, 40, 160, 500, 510,520]

for t in time_steps_to_plot:
    fig, axs = plt.subplots(1, 2, figsize=(20, 6))

    #eta(x, y, t)
    im1 = axs[0].imshow(eta_FT_2d[t, :, :], extent=[0, Xpoints * dx, 0, Ypoints * dy], origin='lower', cmap='RdBu_r', vmin=-0.5, vmax=0.5)
    fig.colorbar(im1, ax=axs[0], label='Amplitude η')
    axs[0].set_xlabel("Position x")
    axs[0].set_ylabel("Position y")
    axs[0].set_title(f"Fig 7 : FTCS waves 2D - η(x, y, t) at t = {t * dt}s")

    # u(x, y, t)
    im2 = axs[1].imshow(u_e_FT_2d[t, 1:, :], extent=[0, Xpoints * dx, 0, Ypoints * dy], origin='lower', cmap='RdBu_r', vmin=-0.1, vmax=0.1)
    fig.colorbar(im2, ax=axs[1], label='Amplitude u')
    axs[1].set_xlabel("Position x")
    axs[1].set_ylabel("Position y")
    axs[1].set_title(f"Fig 8 : FTCS waves 2D - u(x, y, t) at t = {t * dt}s")

    plt.tight_layout()
    plt.show()
```
